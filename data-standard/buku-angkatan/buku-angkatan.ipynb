```python
import pandas as pd
import numpy as np

path = r"/home/al/python-project/data-standard/buku-angkatan/data-nawasena.csv"
data = pd.read_csv(path, sep=";", dtype=str)

data
```

```python
newTable = []

temp = []

for i in range(0,100) :
    row = data.iloc[i,:]
    temp.append(row.tolist())

    if len(temp) == 4 :
        newTable.append(temp)
        temp = []

newTable = np.array(newTable)

flattenTable = []

for i in range(0, 25) :
    flattenTable.append(newTable[i].flatten())
```

```python
newData = pd.DataFrame(columns=["NRP1", "Nama1", "Nomor Telepon1", "Nama Panggilan1", "TTL1", "Asal Kota1", "Hobi1", "NRP2", "Nama2", "Nomor Telepon2", "Nama Panggilan2", "TTL2", "Asal Kota2", "Hobi2", "NRP3", "Nama3", "Nomor Telepon3", "Nama Panggilan3", "TTL3", "Asal Kota3", "Hobi3", "NRP4", "Nama4", "Nomor Telepon4", "Nama Panggilan4", "TTL4", "Asal Kota4", "Hobi4"], data=flattenTable)

newData.to_csv("/home/al/python-project/data-standard/buku-angkatan/buku-nawasena.csv", sep=";", index=False)
```
