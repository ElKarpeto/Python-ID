```python
import pandas as pd

df = pd.read_csv("data_mahasiswa.csv")

group_fakultas = df.groupby("Fakultas")

# group_fakultas.count()
```

```python
data_group = list(group_fakultas.groups.keys())

print(data_group)
```

```python
temp = group_fakultas.get_group(data_group[0]).reset_index(drop=True)

# temp
temp = temp.drop(0).reset_index(drop=True)

temp
```

```python
def countAllJurusan(datas, keys) :
    count = 0
    s = set()
    for key in keys :
        countJurusan = len(datas.get_group(key).groupby("Jurusan"))
        count += countJurusan
    
    return count

data = []

banyakKelompok = 100
banyakOrang = 10

n = len(data_group)

i = 0
index_fakultas = 0
while i < banyakKelompok :
    data_kelompok = []
    s = set()
    j = 0
    index = 0
    if countAllJurusan(group_fakultas, data_group) >= banyakOrang :
        while j < banyakOrang :
            if not group_fakultas.get_group(data_group[index_fakultas]).empty :
                temp = group_fakultas.get_group(data_group[index_fakultas]).reset_index(drop=True)
                if index_fakultas == 1 or temp.iloc[index, 3] not in s :
                    data_kelompok.append(temp.loc[index].to_list())
                    s.add(temp.iloc[index, 3])
                    temp = temp.drop(index).reset_index(drop=True)

                    index_fakultas = (index_fakultas + 1) % n
                    index = 0
                    j += 1
                else :
                    if index >= len(temp) - 1:
                        index_fakultas = (index_fakultas + 1) % n
                        index = 0
                    else :
                        index += 1
            else :
                index_fakultas = (index_fakultas + 1) % n
                index = 0
    else :
        while j < banyakOrang :
            if not group_fakultas.get_group(data_group[index_fakultas]).empty :
                temp = group_fakultas.get_group(data_group[index_fakultas]).reset_index(drop=True)
                data_kelompok.append(temp.loc[index].to_list())
                temp = temp.drop(index).reset_index(drop=True)
                index_fakultas = (index_fakultas + 1) % n
                index = 0
                j += 1
            else :
                index_fakultas = (index_fakultas + 1) % n
                index = 0
    
    data.append(data_kelompok)
    i += 1

for i in range(len(data)) : 
    print(f"Kelompok {i + 1}:")
    for j in range(len(data[0])) :
        print(f"{data[i][j]}")
    
    print("\n")

# for i in range(len(data_group)) :
#     print(f"[debug] : {display(data_group[i])}")

# for i in range(len(data_group)) :
#     print(f"[debug] : {i + 1} = {len(data_group[i])}")
```
